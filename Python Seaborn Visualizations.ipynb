```python
# Import seaborn library

import seaborn as sns
```

```python
# Load the iris data set

iris_data = sns.load_dataset("iris")
```

```python
# Take a look at the data

iris_data.head()
```

```python
# Let's do a scatterplot first

sns.scatterplot(x = "sepal_length", y = "sepal_width", data = iris_data, hue = "species")
```

```python
# Then a histogram

sns.histplot(x = "species", y = "sepal_width", data = iris_data)
```

```python
# How about a bar plot

sns.barplot(x = "species", y = "sepal_width", data = iris_data)
```

```python
# A boxplot will give us a more nuanced look at the sizes

sns.boxplot(x = "species", y = "sepal_width", data = iris_data)
```

```python
# And a violin plot another nuanced look

sns.violinplot(x = "species", y = "sepal_width", data = iris_data)
```

```python
# Let's do a Facet Grid with Distplot for further detail

from warnings import filterwarnings
filterwarnings("ignore")

sns.FacetGrid(iris_data, hue = "species", height = 5).map(sns.distplot, "petal_width").add_legend()
```

```python
# A pair plot will likely provide the most detail and insight into our data set

sns.pairplot(iris_data, hue = "species", height = 3)
```

```python
# A cluster map should provide another interesting look

sns.clustermap(iris_data.drop("species", axis = 1))
```

```python
# Let's check out correlation between our iris features

iris_data.corr
```

```python
# And then use that correlation to create a heatmap

sns.heatmap(iris_data.corr())
```

```python
# Let's try another plot

sns.regplot(x = "sepal_length", y = "sepal_width", data = iris_data, scatter_kws = {"color": "blue"}, line_kws = {"color": "green"})
```
